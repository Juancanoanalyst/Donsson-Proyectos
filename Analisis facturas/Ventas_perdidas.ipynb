```python
import pandas as pd
import datetime
```

```python
df_vp = pd.read_parquet("ventas_perdidas_2025.parquet")
```

```python
df_vp["fecha"] = pd.to_datetime(df_vp["fecha"]) #formato fecha
df_vp["mes"]  = df_vp["fecha"].dt.month #agregar mes
df_vp["ano"]  = df_vp["fecha"].dt.year #agregar año
df_vp["dia"]  = df_vp["fecha"].dt.day #agregar año


df_agosto = df_vp[(df_vp["mes"] == 8)& (df_vp["ano"] == 2025)] #vp durante agosto de 2025

df_agosto_perdidas = df_agosto[df_agosto["cantidad"] > df_agosto["cantidad_existencia"]]

df_agosto_perdidas.info()
```

```python
df_agosto_perdidas.head()
```

```python
df_agosto_perdidas[df_agosto_perdidas["dia"]==26].sort_values(by=["cantidad"],ascending=False).head(5)                                                                                                   
```

```python
df_agosto_perdidas = df_agosto_perdidas.drop(["product_id_num","store_id","warehouse_id","cliente_id","id","cliente_id_num","warehouse_id_num","warehouse_name"],axis=1)
```

```python
df_agosto_perdidas["vp"] = df_agosto_perdidas["cantidad"] - df_agosto_perdidas["cantidad_existencia"] 
```
