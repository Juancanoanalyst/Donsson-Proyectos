### DESVIACION 30%

```python
import pandas as pd

# 4. Calcular coeficientes sobre esas ventas mensuales
coef_por_producto = df.groupby("product_name").agg(
    facturas_unicas = ("invoice_id_num","nunique"), #facturas distintas
    venta_promedio = ("price_subtotal","mean"), # promedio cantidad
    venta_min = ("price_subtotal","min"),
    venta_max =  ("price_subtotal","max")

).reset_index()

coef_por_producto["coeficiente"] = (
    (coef_por_producto["venta_max"] - coef_por_producto["venta_min"]) /
    coef_por_producto["facturas_unicas"]
)

# Clasificaci√≥n por rangos
def clasificar(coef):
    if 0 <= coef < 0.8:
        return "C"
    elif 0.8 <= coef < 1.5:
        return "B"
    elif 1.5 <= coef < 3:
        return "A"
    else:
        return "AAA"

coef_por_producto["Categoria"] = coef_por_producto["coeficiente"].apply(clasificar)

# Ordenar para inspecci√≥n
coef_por_producto = coef_por_producto.sort_values(by="coeficiente", ascending=False)

coef_por_producto.head(30)

#Algo paso esto me qquedo al reves el C deberia Ser AAA
```
